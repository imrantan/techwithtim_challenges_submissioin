```python
# i have defined a function below that will unlock the code for each gucci banana

def gucci_banana_code_unlocker(gucci_banana=str):
    
    #define function to convert the string into a list
    def Convert(string):
        list1=[]
        list1[:0]=string
        return list1
    
    alphabets = Convert('abcdefghijklmnopqrstuvwxyz') #using the convert function defined above to change a string of alphabets into a list
    rank = list(range(1,27)) #get a list of 1 to 26
    dictionary = dict(zip(alphabets, rank)) #using zip to form a dictionary where the rank of the alphabet can be called e.g. 'a': 1 & 'z': 26
    
    x = gucci_banana[::-1] #reverse the string 
    y = "" #initialise an empty string
    
    for num in range(len(x)):
        multiplied_character = dictionary[x[num].lower()]*x[num] #call the rank for the first character in the string and multiply it by the rank
        y = y+multiplied_character #add to the empty string
        
    return y #return the gucci banana!
        
```

```python
gucci_banana_code_unlocker('abc')
```
